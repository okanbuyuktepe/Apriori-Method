```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('D:/Users/okanb/Desktop/jupyter/course_PythonforDataScienceEssentialTrainingPart2/groceries.csv')
data.head()
```

```python
# Format the data as sparse format
# Each transaction is represented by a row
basket_sets = pd.get_dummies(data)
basket_sets.head()
```

### Support Calculation

```python
# Let's take the items that have 0.02 min suppport value
# itemsets -> column index
apriori(basket_sets, min_support=0.02) 
```

```python
# What are these items ?
apriori(basket_sets, min_support=0.02, use_colnames=True)
```

Let's decrease the support values since the support of 0.02 is returning only single popular item

```python
frequent_itemsets = apriori(basket_sets, min_support=0.002, use_colnames=True)

# add a column that describes the length of the combination
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x : len(x))
frequent_itemsets
```

Let's create a filter that where the length is > 3

```python
frequent_itemsets[frequent_itemsets['length'] >= 3]
```

## Association Rules

### Confidence

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)
rules.head()
```

### Lift

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules.head()
```

### Lift and Confidence

```python
rules[(rules['lift'] >=5)& (rules['confidence'] >=.5) ]
```
